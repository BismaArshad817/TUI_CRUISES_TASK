# 1. Analysis of Monthly Power Consumption of different components of Vessel 1 and Vessel 2

````python
# Select relevant columns for power consumption
power_galley_cols = ['Power Galley 1 (MW)', 'Power Galley 2 (MW)', 'Power Service (MW)']
power_hvac_cols = ['HVAC Chiller 1 Power (MW)', 'HVAC Chiller 2 Power (MW)', 'HVAC Chiller 3 Power (MW)']
power_scrubber_col = ['Scrubber Power (MW)']
power_propulsion_cols = ['Propulsion Power (MW)', 'Port Side Propulsion Power (MW)', 'Starboard Side Propulsion Power (MW)']
power_thruster_cols = ['Bow Thruster 1 Power (MW)', 'Bow Thruster 2 Power (MW)', 'Bow Thruster 3 Power (MW)', 
                       'Stern Thruster 1 Power (MW)', 'Stern Thruster 2 Power (MW)']

# Calculate total power consumption for each component
df['Total_Power_Galley'] = df[power_galley_cols].sum(axis=1)
df['Total_Power_HVAC'] = df[power_hvac_cols].sum(axis=1)
df['Total_Power_Scrubber'] = df[power_scrubber_col].sum(axis=1)
df['Total_Power_Propulsion'] = df[power_propulsion_cols].sum(axis=1)
df['Total_Power_Thrusters'] = df[power_thruster_cols].sum(axis=1)

# Extraction of Month from Start Time
df['Start Time'] = pd.to_datetime(df['Start Time'])  
df['Month'] = df['Start Time'].dt.month  

# Group data by Month and Vessel Name to get monthly consumption totals
monthly_data = df.groupby(['Month', 'Vessel Name']).agg(
    Total_Power_Galley=('Total_Power_Galley', 'sum'),
    Total_Power_HVAC=('Total_Power_HVAC', 'sum'),
    Total_Power_Scrubber=('Total_Power_Scrubber', 'sum'),
    Total_Power_Propulsion=('Total_Power_Propulsion', 'sum'),
    Total_Power_Thrusters=('Total_Power_Thrusters', 'sum'),
    Sea_Temperature=('Sea Temperature (Celsius)', 'mean'),
    Speed_Over_Ground=('Speed Over Ground (knots)', 'mean')
).reset_index()

# Define the vessels to plot
vessels = ['Vessel 1', 'Vessel 2']  

for vessel in vessels:
    vessel_data = monthly_data[monthly_data['Vessel Name'] == vessel]
    
# Visualization of power consumption for each component
    plt.figure(figsize=(14, 8))
    
    for column in ['Total_Power_Galley', 'Total_Power_HVAC', 'Total_Power_Scrubber', 
                   'Total_Power_Propulsion', 'Total_Power_Thrusters']:
        plt.plot(vessel_data['Month'].astype(str), vessel_data[column], label=column.replace('_', ' ').title())
    
    plt.title(f'Monthly Power Consumption by Component for {vessel}')
    plt.xlabel('Month')
    plt.ylabel('Power Consumption (MW)')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

````

````python
# Define the power columns based on the actual column names in your DataFrame
power_galley_cols = ['Power Galley 1 (MW)', 'Power Galley 2 (MW)', 'Power Service (MW)']
power_hvac_cols = ['HVAC Chiller 1 Power (MW)', 'HVAC Chiller 2 Power (MW)', 'HVAC Chiller 3 Power (MW)']
power_scrubber_col = ['Scrubber Power (MW)']
power_propulsion_cols = ['Propulsion Power (MW)', 'Port Side Propulsion Power (MW)', 'Starboard Side Propulsion Power (MW)']
power_thruster_cols = ['Bow Thruster 1 Power (MW)', 'Bow Thruster 2 Power (MW)', 'Bow Thruster 3 Power (MW)', 
                       'Stern Thruster 1 Power (MW)', 'Stern Thruster 2 Power (MW)']

# Calculate total power consumption for each component
df['Total_Power_Galley'] = df[power_galley_cols].sum(axis=1)
df['Total_Power_HVAC'] = df[power_hvac_cols].sum(axis=1)
df['Total_Power_Scrubber'] = df[power_scrubber_col].sum(axis=1)
df['Total_Power_Propulsion'] = df[power_propulsion_cols].sum(axis=1)
df['Total_Power_Thrusters'] = df[power_thruster_cols].sum(axis=1)

# Group by vessel name and calculate total power for each component
vessel_power_consumption = df.groupby('Vessel Name').agg(
    Total_Power_Galley=('Total_Power_Galley', 'sum'),
    Total_Power_HVAC=('Total_Power_HVAC', 'sum'),
    Total_Power_Scrubber=('Total_Power_Scrubber', 'sum'),
    Total_Power_Propulsion=('Total_Power_Propulsion', 'sum'),
    Total_Power_Thrusters=('Total_Power_Thrusters', 'sum')
).reset_index()

# Create pie charts for each vessel
for index, row in vessel_power_consumption.iterrows():
    total_power_consumption = {
        'Total Power Galley': row['Total_Power_Galley'],
        'Total Power HVAC': row['Total_Power_HVAC'],
        'Total Power Scrubber': row['Total_Power_Scrubber'],
        'Total Power Propulsion': row['Total_Power_Propulsion'],
        'Total Power Thrusters': row['Total_Power_Thrusters']
    }

# Convert to a DataFrame for easier plotting
    total_power_df = pd.DataFrame(total_power_consumption.items(), columns=['Component', 'Total Power (MW)'])

# Visualization of Pie Chart
    plt.figure(figsize=(10, 10))
    plt.pie(total_power_df['Total Power (MW)'], labels=total_power_df['Component'], autopct='%1.1f%%', startangle=140)
    plt.title(f'Total Power Consumption by Component for Vessel: {row["Vessel Name"]}')
    plt.axis('equal')  
    plt.show()
````

# 1. Total Diesel Generator Power Generation (MW)

````python
# Function to calculate the total power generation for diesel generators for a specific vessel
def calculate_power_generation(data, vessel_name):
    vessel_data = data[data['Vessel Name'] == vessel_name]
    diesel_power = {
        'Diesel Generator 1': vessel_data['Diesel Generator 1 Power (MW)'].sum(),
        'Diesel Generator 2': vessel_data['Diesel Generator 2 Power (MW)'].sum(),
        'Diesel Generator 3': vessel_data['Diesel Generator 3 Power (MW)'].sum(),
        'Diesel Generator 4': vessel_data['Diesel Generator 4 Power (MW)'].sum()
    }
    return diesel_power

vessel_1 = 'Vessel 1'  
vessel_2 = 'Vessel 2'  

power_generation_vessel_1 = calculate_power_generation(df, vessel_1)
power_generation_vessel_2 = calculate_power_generation(df, vessel_2)

# Visualization of two different plots for both vessels
def plot_power_generation(vessel_1_data, vessel_2_data, vessel_1_name, vessel_2_name):
    bar_width = 0.35
    index = range(len(vessel_1_data))

    plt.figure(figsize=(12, 6))
    bars1 = plt.bar(index, vessel_1_data.values(), bar_width, label=vessel_1_name, color='blue')
    bars2 = plt.bar([i + bar_width for i in index], vessel_2_data.values(), bar_width, label=vessel_2_name, color='orange')

    plt.xlabel('Diesel Generators')
    plt.ylabel('Power Generated (MW)')
    plt.title('Power Generation by Diesel Generators for Two Vessels')
    plt.xticks([i + bar_width / 2 for i in index], vessel_1_data.keys())
    plt.legend()
    plt.ylim(0, max(max(vessel_1_data.values()), max(vessel_2_data.values())) * 1.1)  
    plt.grid(axis='y')

    for bars in [bars1, bars2]:
        for bar in bars:
            plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.2f}', ha='center', va='bottom')

    plt.tight_layout()
    plt.show()

# Call the plotting function for both vessels
plot_power_generation(power_generation_vessel_1, power_generation_vessel_2, vessel_1, vessel_2)






````

# 1. Monthly Power Consumption and Generation Analysis KPI

````python
# Extract month and year from 'Start Time'
df['Month'] = df['Start Time'].dt.to_period('M')

# Select relevant columns for power consumption and generation
power_consumption_cols = [
    'Power Galley 1 (MW)', 'Power Galley 2 (MW)', 'Power Service (MW)',
    'HVAC Chiller 1 Power (MW)', 'HVAC Chiller 2 Power (MW)', 
    'HVAC Chiller 3 Power (MW)', 'Scrubber Power (MW)', 
    'Propulsion Power (MW)', 'Port Side Propulsion Power (MW)', 
    'Starboard Side Propulsion Power (MW)', 'Bow Thruster 1 Power (MW)', 
    'Bow Thruster 2 Power (MW)', 'Bow Thruster 3 Power (MW)', 
    'Stern Thruster 1 Power (MW)', 'Stern Thruster 2 Power (MW)'
]

power_generation_cols = [
    'Diesel Generator 1 Power (MW)', 'Diesel Generator 2 Power (MW)', 
    'Diesel Generator 3 Power (MW)', 'Diesel Generator 4 Power (MW)'
]

# Calculate total power consumption and generation for each row
df['Total_Power_Consumption'] = df[power_consumption_cols].sum(axis=1)
df['Total_Power_Generation'] = df[power_generation_cols].sum(axis=1)

# Calculate monthly sums for power consumption and generation per vessel
monthly_data = df.groupby(['Month', 'Vessel Name']).agg(
    Power_Consumption=('Total_Power_Consumption', 'sum'),
    Power_Generation=('Total_Power_Generation', 'sum')
).reset_index()

# Plotting for each vessel separately
vessels = monthly_data['Vessel Name'].unique()

for vessel in vessels:
    vessel_data = monthly_data[monthly_data['Vessel Name'] == vessel].copy()  
    fig, ax = plt.subplots(figsize=(12, 6))
    bar_width = 0.35
    vessel_data.loc[:, 'Month_str'] = vessel_data['Month'].astype(str)  
    
    ax.bar(vessel_data['Month_str'], vessel_data['Power_Consumption'], 
           width=bar_width, label='Total Power Consumption (MW)', alpha=0.7)
    ax.bar(vessel_data['Month_str'], vessel_data['Power_Generation'], 
           width=bar_width, label='Total Power Generation (MW)', alpha=0.7, bottom=vessel_data['Power_Consumption'])

    ax.set_xlabel('Month')
    ax.set_ylabel('Power (MW)')
    ax.set_title(f'Monthly Power Consumption vs Power Generation for {vessel}')
    ax.set_xticks(vessel_data['Month_str'])
    ax.set_xticklabels(vessel_data['Month_str'], rotation=45)
    ax.legend()

    plt.tight_layout()
    plt.show()
````

# 2. Energy Utilization Efficiency KPI (efficiency of total power generation over the total fuel consumption)

````python
# Extract month and year from 'Start Time'
df['Month'] = df['Start Time'].dt.to_period('M')

# Select relevant columns for power consumption and generation
power_consumption_cols = [
    'Power Galley 1 (MW)', 'Power Galley 2 (MW)', 'Power Service (MW)',
    'HVAC Chiller 1 Power (MW)', 'HVAC Chiller 2 Power (MW)', 
    'HVAC Chiller 3 Power (MW)', 'Scrubber Power (MW)', 
    'Propulsion Power (MW)', 'Port Side Propulsion Power (MW)', 
    'Starboard Side Propulsion Power (MW)', 'Bow Thruster 1 Power (MW)', 
    'Bow Thruster 2 Power (MW)', 'Bow Thruster 3 Power (MW)', 
    'Stern Thruster 1 Power (MW)', 'Stern Thruster 2 Power (MW)'
]

power_generation_cols = [
    'Diesel Generator 1 Power (MW)', 'Diesel Generator 2 Power (MW)', 
    'Diesel Generator 3 Power (MW)', 'Diesel Generator 4 Power (MW)'
]

# Calculate total power consumption and generation for each row
df['Total_Power_Consumption'] = df[power_consumption_cols].sum(axis=1)
df['Total_Power_Generation'] = df[power_generation_cols].sum(axis=1)

# Calculate monthly sums for power consumption and generation per vessel
monthly_data = df.groupby(['Month', 'Vessel Name']).agg(
    Power_Consumption=('Total_Power_Consumption', 'sum'),
    Power_Generation=('Total_Power_Generation', 'sum')
).reset_index()

# Create Month_str column
monthly_data['Month_str'] = monthly_data['Month'].astype(str)

# Calculate Power Efficiency Ratio
monthly_data['Power_Efficiency_Ratio'] = monthly_data['Power_Generation'] / monthly_data['Power_Consumption']

# Visualizing Power Efficiency Ratio
plt.figure(figsize=(12, 6))
for vessel in monthly_data['Vessel Name'].unique():
    vessel_data = monthly_data[monthly_data['Vessel Name'] == vessel]
    plt.plot(vessel_data['Month_str'], vessel_data['Power_Efficiency_Ratio'], label=vessel)

plt.xlabel('Month')
plt.ylabel('Power Efficiency Ratio')
plt.title('Power Efficiency Ratio by Vessel')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

# 3. Energy Efficiency Index KPI (Efficiency in energy usage, incorporating speed as a factor)

````python
# Calculate monthly sums for power consumption per vessel
monthly_data = df.groupby(['Month', 'Vessel Name']).agg(
    Power_Consumption=('Total_Power_Consumption', 'sum'),
    Speed_Over_Ground=('Speed Over Ground (knots)', 'mean')  
).reset_index()

# Calculate Energy Efficiency Index 
monthly_data['Specific_Power_Consumption'] = monthly_data['Power_Consumption'] / monthly_data['Speed_Over_Ground']

# Visualize Energy Efficiency Index 
plt.figure(figsize=(12, 6))
for vessel in monthly_data['Vessel Name'].unique():
    vessel_data = monthly_data[monthly_data['Vessel Name'] == vessel]
    plt.plot(vessel_data['Month'].astype(str), vessel_data['Specific_Power_Consumption'], label=vessel)


plt.xlabel('Month')
plt.ylabel('Specific Power Consumption (MW/knot)')
plt.title('Specific Power Consumption Efficiency Relative to Speed')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

# 4. Energy Efficiency Ratio KPI

````python
# Calculate total power generation from diesel generators
df['Total_Power_Generation'] = (
    df['Diesel Generator 1 Power (MW)'] +
    df['Diesel Generator 2 Power (MW)'] +
    df['Diesel Generator 3 Power (MW)'] +
    df['Diesel Generator 4 Power (MW)']
)

# Calculate total fuel consumption
df['Total_Fuel_Consumption'] = (
    df['Main Engine 1 Fuel Flow Rate (kg/h)'] +
    df['Main Engine 2 Fuel Flow Rate (kg/h)'] +
    df['Main Engine 3 Fuel Flow Rate (kg/h)'] +
    df['Main Engine 4 Fuel Flow Rate (kg/h)'] +
    df['Boiler 1 Fuel Flow Rate (L/h)'] * 0.85 +  
    df['Boiler 2 Fuel Flow Rate (L/h)'] * 0.85 + 
    df['Incinerator 1 Fuel Flow Rate (L/h)'] * 0.85  
)

# Calculate Energy Efficiency Ratio (EER)
df['Energy_Efficiency_Ratio'] = df['Total_Power_Generation'] / df['Total_Fuel_Consumption']

# Group by month and vessel name to analyze EER over time
monthly_eer = df.groupby(['Month', 'Vessel Name']).agg(
    Average_EER=('Energy_Efficiency_Ratio', 'mean'),
    Total_Power_Generation=('Total_Power_Generation', 'sum'),
    Total_Fuel_Consumption=('Total_Fuel_Consumption', 'sum')
).reset_index()

# Visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
for vessel in monthly_eer['Vessel Name'].unique():
    vessel_data = monthly_eer[monthly_eer['Vessel Name'] == vessel]
    plt.plot(vessel_data['Month'].astype(str), vessel_data['Average_EER'], label=vessel)

plt.xlabel('Month')
plt.ylabel('Average Energy Efficiency Ratio (MW/kg/h)')
plt.title('Energy Efficiency Ratio by Vessel')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

# 5. Fuel Efficiency Ratio KPI (Fuel Consumption relative to Distance Travelled)

````python
# Calculate total fuel consumption in kilograms
df['Total_Fuel_Consumption'] = (
    df['Main Engine 1 Fuel Flow Rate (kg/h)'] +
    df['Main Engine 2 Fuel Flow Rate (kg/h)'] +
    df['Main Engine 3 Fuel Flow Rate (kg/h)'] +
    df['Main Engine 4 Fuel Flow Rate (kg/h)']
)

# Calculate Fuel Efficiency Ratio 
df['Fuel_Efficiency_Ratio'] = df['Distance (nautical miles)'] / df['Total_Fuel_Consumption']

# Group by month and vessel name to analyze FER over time
monthly_fer = df.groupby(['Month', 'Vessel Name']).agg(
    Average_FER=('Fuel_Efficiency_Ratio', 'mean'),
    Total_Fuel_Consumption=('Total_Fuel_Consumption', 'sum'),
    Total_Distance=('Distance (nautical miles)', 'sum')
).reset_index()

# Visualization
plt.figure(figsize=(12, 6))
for vessel in monthly_fer['Vessel Name'].unique():
    vessel_data = monthly_fer[monthly_fer['Vessel Name'] == vessel]
    plt.plot(vessel_data['Month'].astype(str), vessel_data['Average_FER'], label=vessel)

plt.xlabel('Month')
plt.ylabel('Fuel Efficiency Ratio (nm/kg)')
plt.title('Fuel Efficiency Ratio by Vessel')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````
